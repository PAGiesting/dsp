The numbers generated by random.random are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.
Generate 1000 numbers from random.random and plot their PMF and CDF. Is the distribution uniform?

```python
import numpy as np
import thinkstats2 as ts2
import thinkplot as tp

sample = np.random.random(1000)
sample[995:999]
```

```python
sample_cdf = ts2.Cdf(sample)
```

```python
tp.Cdf(sample_cdf)
tp.Show(xlabel='value', ylabel='cumulative')
```

```python
# Let's see how badly this breaks.

sample_pmf = ts2.Pmf(sample)
tp.Pmf(sample_pmf)
tp.Show(xlabel='value', ylabel='probability')
```

```python
sample_hist = ts2.Hist(sample)
tp.Hist(sample_hist)
tp.Show(xlabel='value', ylabel='counts')
```

That didn't work at all.
The CDF shows a few waggles in the expected straight line.
The unbinned PMF is basically useless, and the likewise unbinned histogram really is useless.
From the CDF, I would guess that the random number generator is doing an ok job. I would expect that as I increased the number of random choices, the CDF should get closer to a straight line. Let me see if I can break this little MacBook.

```python
sample6 = np.random.random(1000000)
sample6_cdf = ts2.Cdf(sample6)
tp.Cdf(sample6_cdf)
tp.Show(xlabel='value', ylabel='cumulative')
```

Hey, that was easy enough, and the plot is completely straight to the eye / pixel.
